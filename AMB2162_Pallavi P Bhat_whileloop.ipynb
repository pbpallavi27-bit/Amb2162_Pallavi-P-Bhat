```python
# add numbers from 1 until the sum becomes greater than 50
i=1
current_sum=0
# The loop should continue as long as current_sum is less than or equal to 50
while current_sum <= 50:
  current_sum += i # Add the current number 'i' to the sum
  i += 1           # Increment 'i' for the next iteration
# After the loop, current_sum will be the first sum that is greater than 50
print(current_sum)

```
