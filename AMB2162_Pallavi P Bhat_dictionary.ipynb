```python
# give me the maximum valued key as an output
# give me a key which has maximum value
CGPA={1:9.1,2:8.1,3:8.7,4:6.8,6:7.9}
print("key with maximum value-->",max(CGPA, key=CGPA.get))#GEMINI
x= zip(CGPA.values(),CGPA.keys())
max_cgpa = max(x) # storing maximum in another variable so you dont get zip object not subscriptable
print("the maximum CGPA with its key respectively is", max_cgpa)
print("The key for the maximum CGPA is", max_cgpa[1]) # Access the key from the stored pair
```

```python
#method 1 to find key for min value
CGPA={1:9.1,2:8.1,3:8.7,4:6.8,6:7.9}
keylist=list(CGPA.keys())
vallist=list(CGPA.values())
minval=min(CGPA.values())
print("\nValue=",minval,"key-->",keylist[vallist.index(minval)])
```

```python
#method 1 to find key for max value
CGPA={1:9.1,2:8.1,3:8.7,4:6.8,6:7.9}
keylist=list(CGPA.keys())
vallist=list(CGPA.values())
maxval=max(CGPA.values())
print("\nValue=",maxval,"key-->",keylist[vallist.index(maxval)])
```

```python
#copying from one dictionary tp another
# method no 1
dict1={"name":"John",10:[2,4,3]}
dict2=dict1
print(dict1)
print(dict2)
print()
print(" after doing change")
print() # for space
dict1.popitem()
print(dict1)
print(dict2)
```

```python
#copying from one dictionary tp another
# method no 2 copy method
dict1={"name":"John",10:[2,4,3]}
dict2=dict1.copy()
print(dict1)
print(dict2)
print()
print(" after doing change")
print() # for space
dict1.popitem()
print(dict1)
print(dict2)
print("change in dict1 will not reflect in dict 2")
```
