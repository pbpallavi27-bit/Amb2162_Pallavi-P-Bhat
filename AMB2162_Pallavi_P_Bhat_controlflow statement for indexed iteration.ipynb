```python
#CREATE A list of soil moisture levels in percentage
soil_moisture=[45,60,30,55,40]
#display the values in the format %
print("soil moisture levels in different fields-")
for moisture in soil_moisture:
  print(moisture,"%")
```

```python
#3 if moisture <50 display"irrigation needed" else display as "sufficient moisture"
for moisture in soil_moisture:
  if moisture<50:
    print(moisture,"%", "irrigation needed")
  else:
    print(moisture,"%", "sufficient moisture")
```

```python
#4 display total number of fields tested(length)
print("total number of field checked",len(soil_moisture))
```

```python
#5 add new moisture value with 65 and check the count
soil_moisture=[45,60,30,55,40]
soil_moisture.append(65)
print("after adding new field moisture:")
print(soil_moisture)
print("total fields checked:",len(soil_moisture))
```

```python
#6 remove 30 and check the count
soil_moisture=[45,60,30,55,40]
soil_moisture.remove(30)
print("after removing 30% moisture value:")
print(soil_moisture)
print(" total fields checked:",len(soil_moisture))
```

```python
#7 sort the values
soil_moisture.sort()
print(soil_moisture)
```

```python
# 8 print maximum and minimum of the moisture value
print("minimum moisture:",min(soil_moisture),"%")
print("maximum moisture:",max(soil_moisture),"%")
```

```python
#using range and length
crops=["Rice","Wheat","Millet","Maize","Pulses"]
water_needed=[300,200,100,150,80] #in mm
print("length",len(crops))
print("range",range(len(crops)))
print("CROP AND RESPECTIVE WATER REQUIREMENT:")
for i in range(len(crops)):
  print(crops[i],"-",water_needed[i],"mm")
```

```python
#using zip
print("CROP AND THEIR WATER REQUIREMENT")
for crop,water in zip(crops,water_needed):
  print(crop,"-",water,"mm")
```

```python
#if in for .....checking the condition
print(" CROPS NEEDING HIGH WATER >180mm :")
for i in range(len(water_needed)):
  if water_needed[i]>180:
    print(crops[i])
```

```python
#if in zip...checking the condition
print(" CROPS REQUIRING HIGH WATER> 180mm:")
for crop,water in zip(crops,water_needed):
  if water>180: # don't use [i] here
     print(crop)
```

```python
crops=["rice","maize","sugarcane","cotton","pulses"]
water_need=[1250,500,2200,600,320]
index=water_need.index(max(water_need))
print("crop with high water requirement:",crops[index])
```

```python
#1 method
crops=["rice","wheat","maize","pulses"]
yield_tons=[3.5,4.0,2.8,1.5]

index=yield_tons.index(max(yield_tons))
print("highest yielding crop:",crops[index])
```
