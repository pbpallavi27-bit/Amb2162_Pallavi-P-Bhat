```python
#read the soil pH and interpret the values;
a=float(input(" ENTER YOUR SOIL PH "))
if(a<=5.5):
  print( "SOIL IS ACIDIC")
elif(a>5.5 and a<=7.5):
  print(" SOIL IS NUETRAL")
elif(a>7.5 and a<=8.5):
  print( "SOIL IS ALKALINE ")
else:
  print(" SOIL IS HIGHLY ALKALINE ")
```

```python
# is called comment
#create a list with cops rice wheat maize and display the crops using for loop
crops=["Rice","Wheat","Maize"]
#display crops
print("crops grown in the field :")
for i in crops:
  print(i)
#i means irrirerior value
# add sugarcane to list
crops.append("Sugarcane")
print(crops)
#display numbe of crops
print("number of crops grown:",len(crops))
#check if maize and onion are present
if "Maize" in crops:
  print("maize is present in the list of crops")
if"onion" in crops:
  print(" onion is present in list of crops ")
else:
  print("onion is not present in the list of crops")
#remove the wheat and find the count
crops.remove("Wheat")
print(crops)
print(len(crops))
#find index of a crop
index_maize=crops.index("Maize")
print("Maize is at index:",index_maize)
print(crops.index("Maize"))
#insert a new crop at specific position
crops.insert(2,"onion")
print(crops)
#remove the last crop
crops.pop()
print("after removing the last crop")
print(crops)
#sort the crops
crops.sort()
print(crops)
#reverse the list
crops.reverse()
print(crops)
```
