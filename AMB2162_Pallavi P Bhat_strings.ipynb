```python
# assign a variable with "Today is 19-Jan-2026 MONDAY"
#calculate number of vowels , consonants,blankspaces
#and other characters
xx="Today is 19-Jan-2026 MONDAY"
b=v=c=d=o=0
for i in xx:
  if i==" ":
    b=b+1
  elif i in "aeiou":
    v=v+1
  elif i.isalpha():
    c=c+1
  elif i.isdigit():
    d=d+1
  else:
    o=o+1
print(  'number of blank spaces=', b)
print( "number of vowels in xx=", v)
print(  "number of consonants in xx=", c)
print( "number of digits in xx=",d)
print( "number of other characters =", o)
```
