```python
list1=[1,2,4,5.56,'jarvis',3]#heterogeneous elements
print(list1)
print(len(list1))#len tells the lenghth of the list
print(list1.index(3))
list1.append('hi')#append adds value at last
print(list1)
print(list1.count('hi'))#counts the value
print(list1.count(76))
list1.insert(2,'hello')#inserts the value at the specific index  (index)(value)
print(list1)
list1.reverse()
print(list1)
list1.pop()#last element is removed
print(list1)
list1.pop(3)#pop(index)
print(list1)
list1.remove('hello')#removes specified value
print(list1)
list1.clear()#removes all elements from the list
print(list1)
#i want to completely removes the list from the memory disc del
del(list1)
```
