```python
# create a dictionary
fld={"rice":120 ,"wheat":85, "maize":60 }
print(fld)
```

```python
#1 find the number of crops cultivated by the farmer
print("number of crops cultivated by farmer=",len(fld.keys()))
```

```python
#2 display the yield of wheat in 2 methods get and without get method
print("yield of wheat is ",fld["wheat"],"quintals per acre")
```

```python
#2 display the yield of wheat in 2 methods get and without get method
print("yield of wheat in get method=",fld.get("wheat"))
```

```python
#3 add a crop pulses yield 45
fld["pulses"]=45
print("after adding pulses",fld)
```

```python
#5 update the yield of rice to 130
fld["rice"]=130
print("after updating",fld)
```

```python
#6 display all crop names cultivated by farmer
print("name of all the crops cultivated by farmer =",list(fld.keys()))
for  i in fld :
   print(i)
```

```python
#7 display all crop yield ( dont want dict._)
print("yield of all the crops cultivated by farmer =",list(fld.values()))
```

```python
#8 check wheather sugarcane is cultivated by the farmer
print("sugar cane is cultivated by farmer =","sugarcane"in fld)
```

```python
#9 increase the yield of wheat by 10
fld["wheat"] = fld["wheat"] + 10
print(fld)
```

```python
#10 caluclate total yield of all crops
print("total yield of all crops=",sum(fld.values()),"quintals/acre")
```

```python
#11 display each crop and its yield using loop
for  i in fld :
   print(i,fld[i])
```

```python
for k,v in fld.items():#  items is added here
  print(k,v)
```

```python
#12 copy the dictionary print the values of new dictionary
fld1=fld.copy()
print(fld1)
```

```python
#13 clear the elements from the new dictionary
fld1.clear()
print("the original ",fld)
print("copied version",fld1)
```
