```python
#before
numbers=[10,20,30,40,50,60]
n=[]
for i in numbers:
  n.append(i+2)
  print(n)
```

```python
#list comprehension
numbers=[10,20,30,40,50,60]
n=[i+2 for i in numbers]
print("new list",n)
```

```python
# 2 create a new list divisible by 4
n=[]
for i in numbers:
  if i%4==0:
    n.append(i)
    print("numbes divisible by 4",n)
```

```python
# using list comprehension
numbers=[10,20,30,40,50,60]
f=[i for i in numbers if i%4==0]
print(" numbers divisible by 4",f)
print("original list",numbers)
```

```python
# even or odd using list comprehension and if else
numbers=[10,21,30,41,50,61]
n=["even" if i%2==0 else"odd" for i in numbers]
print(n)
```
