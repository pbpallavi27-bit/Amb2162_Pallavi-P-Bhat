```python
def expr(v,f):
  try:
    c=v/f #a/b
  except ZeroDivisionError:
        print("divided by 0")
  except NameError:
        print("variable not declared")
  except:
       print("some error occured")
  else:
       print("executed without erors")
       print("value of c",c)
  finally:
       print("end of the program")

v=int(input("enter a dividend"))
f=int(input("enter a divisor"))
expr(v,f)
```

```python
def vending_machine(money, item):
    prices = {"chips": 2, "soda": 1.5}
    if money >= prices.get(item, 0):
        return f"Here is your {item}!"
    else:
        return "Not enough money."

print(vending_machine(100, "lays"))  # Output: Here is your chips!
#Your use of prices.get(item, 0) creates a loophole.

#The .get() method returns 0 if the item is not in the dictionary.
#Therefore, if you ask for an item that doesn't exist (like "pizza") and offer 0 money, the condition 0 >= 0 evaluates to True.
```
