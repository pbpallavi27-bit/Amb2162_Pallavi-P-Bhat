```python
import pandas as pd
import numpy as np
# create data with missing value
data=pd.DataFrame({'plot':['A','B','C','D','E'],'yield_kg':[2500,np.nan,2800,np.nan,2600],
                   'rainfall':[150,200,np.nan,180,170],'TEMPERATURE':[28,30,29,np.nan,27]})
print("===data with missing value===")
print("is null")
print(data.isnull())
print(f"\ntotal missing per coloumn")
print(data.isnull().sum())
#nan means not a number
```

```python
# ===filling missing value ===
print("\n===Filling strategies ===")
#fill with mean
data_mean=data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg']=data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean)
print("\n fill with mean:")
print(data_mean['yield_kg'])
```

```python
# ===filling missing value ===
print("\n===Filling strategies ===")
#fill with mean
data_mean=data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg']=data_mean['yield_kg'].fillna(data_mean['yield_kg'].median)
print("\n fill with median:")
print(data_mean['yield_kg'])
```

```python
# forward fill (use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print('\n forward fill')
print(data_ffill)
```

```python
#==dropping the missing data
print("\n>>>dropping the rows<<<")
data_dropped=data.dropna()
print(data_dropped)
print(f"original rows :{len(data)},after drop :{len(data_dropped)}")
```
