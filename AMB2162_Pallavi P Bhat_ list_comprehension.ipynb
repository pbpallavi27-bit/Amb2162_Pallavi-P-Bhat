```python
#1 identify the most effective pesticide and # 2 least effective
pesticides=["Neem oil","Chlorpyriphos","Malathion","Imidacloprid","Spinosad"]
pst_eft=[75,85,70,90,88]
mosteffective_pesticide=max(zip(pst_eft,pesticides))
print(mosteffective_pesticide)
print(" MOST EFFECTIVE PESTICIDE AMONG THE FOLLOWING IS-",mosteffective_pesticide[1])
lst_eft=min(zip(pst_eft,pesticides))
print(lst_eft)
print(" LEAST EFFECTIVE PESTICIDE AMONG THE FOLLOWING IS-",lst_eft[1])
```

```python
# 3 average effectiveness percentage

avg_eft=sum(pst_eft)/len(pst_eft)
print(" AVERAGE EFFECTIVENESS % IS ",avg_eft)
```
