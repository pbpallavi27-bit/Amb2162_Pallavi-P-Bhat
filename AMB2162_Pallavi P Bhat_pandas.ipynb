```python
import pandas as pd
production = pd.Series([30,60,45,25],index=["wheat","rice","maize","barley"])
print(production)
```

```python
#by crop name label
print(production["rice"])
print(production.loc["rice"]) #location by label
print(production.iloc[2]) # location by indexing
print(production.sum())
print(production.mean())
print(production.max())
print(production.idxmax()) #maximum production crop
# it prints the index label of the maximum value in the series
```

```python
# crops with production greater than 40 tons
high_production=production[production>40] # inner one gives true or false , outer variable production
# is used for name of that
print(high_production)
print()# for space

print(production>40)
```

```python
# a data frame is a two dimentional data structure.a table like excel sheet -made of multiple series
import pandas as pd
data={
    "Farm":["farm a","farm b","farm c","farm d"],
     "crop":["wheat","rice","maize","redgram"],
    "area_hectares":[10,14,12,3],
    "production_tons":[12,22,31,7]
}

df=pd.DataFrame(data)
print(df)
df.head(2) # to get top 2
df.tail(2) # bottom row
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.describe() #summary statistics
```

```python
#add a new coloumn to existing one
df['new_coloumn_name']=df["production_tons"]/df["area_hectares"]
print(df)
#data frame is permanently changed

```

```python
#display only farm d and farm d
df.iloc[[1,3]]
```

```python
#slicing of the coloumn
df.iloc
```

```python
#filter data
#farms growing wheat
print(df["crop"]=="wheat")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
# sort farms by high production
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
#group by operation
avg_yield=df.groupby("crop")["new_coloumn_name"].sum()
print(avg_yield)
```
