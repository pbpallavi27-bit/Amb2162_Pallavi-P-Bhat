```python
# crop yields for 4 farms over 5 seasons (in tons )
import numpy as np
crop_yields=np.array([
   [12.5,13.2,14.1,13.8,15.2],
   [10.3,11.5,10.8,12.1,11.9],
   [15.1,14.8,16.2,15.5,16.8],
   [11.2,10.9,12.3,11.7,12.8]
 ])

print("crop yields array(tons):")
print(crop_yields)
print(f'\n shape:{crop_yields.shape}')
print(f'\n data type:{crop_yields.dtype}')
print(f'\n total number of elements:{crop_yields.size}')
print(f'\n number of dimentions:{crop_yields.ndim}')
```

```python
# agricultural analysis operations
total_yield =np.sum(crop_yields)
avg_yield_per_farm=np.mean(crop_yields,axis=1) #axis acts on row vise
avg_yield_per_season=np.mean(crop_yields,axis=0)
max_yield=np.max(crop_yields)
min_yield=np.min(crop_yields)

print(f'\n Total production:{total_yield}tons')
print(f'\n average per farm:{avg_yield_per_farm}tons')
print(f'\n average per season:{avg_yield_per_season}tons')
print(f'\n maximum yield :{max_yield}tons')
print(f'\n minimum yield:{min_yield}tons')
```

```python
#find best performing farm
best_farm=np.argmax(avg_yield_per_farm)
#np.argmax()does not return the max value itself _ it returns where that value is located
print("\n best performing farm:farm",best_farm+1)
print("their average yield ",round(avg_yield_per_farm[best_farm],2),"tons")

#or
print("\n best performing farm:farm,{best_farm+1}")
print("their average yield :{avg_yield_per_farm[best_farm]:.2},tons")
# if you take max it gives among 20 elements
# if you want to find it in farm 1 you have to give axis=1
```

```python
# boolean indexing finding exceptional yields
exceptional=crop_yields>15
print(exceptional)
print('\n exceptional yields (>15):')
print(crop_yields[exceptional])
```
