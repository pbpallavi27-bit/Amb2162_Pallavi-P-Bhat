```python
 # WELCOME TO THE REALM OF "SETS!!"
S={"E","A","E","B","F","E","F"}
print("S-->",S)
print("number of elements in set-->",len(S))
```

```python
#type
print("type of object-->",type(S))
```

```python
#traverse set elements
#traverse means travel across
S={"E","A","E","B","F","E","F"}
print(S)
print("")
for i in S:
 print(i)
```

```python
# set elements are unindexed
S={"E","A","E","B","F","E","F"}
print(S)
print(S[2])
```

```python
#adding elements to set using "add"
S={"E","A","E","B","F","E","F"}
print(S)
S.add("P")
S.add("P")
print("after adding an element",S)
# by adding again there is no change in the set
S.add("C")
print(S)

```

```python
#Adding elements to set using "update"
#once the set is created we cannot change the items , but can add items
S1={"p","a","l","l","a","v","i"}
S2={"p","a","v","t","h","r","a"}
print("S1",S1)
print("S2",S2)
S1.update(S2) # inserts the items of S2 into S1 that is Union of two sets
print("updated S1",S1)
```

```python
#removing the elements from the set using remove()
S={"E","A","E","B","F","E","F"}
print("the original",S)
S.remove("E")
#s.remove(p) raises an exception and shows error
print("after removing",S)
```

```python
#removing element from the set using discard()
S={"E","A","E","B","F","E","F"}
print("the original",S)
S.discard("E")
S.discard("P")
# this does not show any error for s.discard(p)
print("after removing",S)
```

```python
#removing element from the set using pop
S={"E","A","E","B","F","E","F"}
print("the original",S)
S.pop()
S.pop()
print("poped",S)
#if the set is empty pop() will raise an exception
```

```python
#clear elements in set (deleting all elements)
S.clear()
print("after clearing the set S-->",S)
```

```python
#deleting the set
del(S)
#del(set name)
```

```python
# sum and average of the sets
S={21,27,12,7,23,15}
print("sum of the elements in the set=",sum(S))
```

```python
#average
print("average of the elements in the set=",sum(S)/len(S))
```

```python
# max and min
print("maximum value element in the set=",max(S))
print("minimum value element in the set=",min(S))
```

```python
# finding the presence of an element in the set
S={21,27,12,7,23,15}
print("27 is present ",27 in S)
print("25 is present ",25 in S)
```

```python
#operations on sets
#intersection , union, difference,symetric difference
a={21,27,15,62,68,77,7}
b={2,3,4,21,7,9}
print("union", a.union(b))
print("intersection", a.intersection(b))
print("difference" , a.difference(b))
print("b-a=",  b-a)
print("a-b =", a-b)
print(a.symmetric_difference(b))
print(b.symmetric_difference(a))
#The symmetric difference of two sets (A △ B) is the collection of elements
# that are in either set A or set B, but not in both (their intersection)
```
