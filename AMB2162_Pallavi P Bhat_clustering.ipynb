```python
# Reading the dataset
dataset=pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:,[3,4]].values


from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
y_kmeans


len(y_kmeans)


print(np.round(kmeans.cluster_centers_[:,0],3))
print(np.round(kmeans.cluster_centers_[:,1],3))


plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='teal',label='Centroids')
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending score(1-100)')
plt.show()
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],c='green',label='Cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],c='blue',label='Cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],c='cyan',label='Cluster 4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],c='magenta',label='Cluster 5')
plt.scatter(X[y_kmeans==5,0],X[y_kmeans==5,1],c='pink',label='Cluster 6')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='teal',label='Centroids')
plt.legend()
plt.title('Clusters of customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending score(1-100)')
plt.show()






```

```python
#Elbow method to find the optimal number of clusters
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

print(wcss)
```
