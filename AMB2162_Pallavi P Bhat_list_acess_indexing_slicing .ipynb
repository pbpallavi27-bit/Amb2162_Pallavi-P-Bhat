```python
#slicing
list1=[10,20,30,40,50,60]
print(list1[0])
print(list1[2:4])
print(list1[3:6])
print(list1[2:8])
print(list1[:3])
print(list1[3:])
```

```python
# 2 types of indexing
#forward indexing and backward indexing
list1=[10,20,30,40,50,60]
     # -6,-5,-4,-3,-2,-1
print(list1[2])
print(list1[-2])
print(list1[1:6])
print(list1[1:6:2])
print(list1[1:6:1])
print(list1[1:6:])
print(list1[1:6:3])
```

```python
#basic list slicing
#syntax : list[start:end:step]
numbers=[10,20,30,40,50,60]
       # -6,-5,-4,-3,-2,-1
print(numbers[-5:3:-1])
print(numbers[-5:-1:2])
```

```python
crops=["rice","wheat","maize","millet","pulses"]
print(crops[:3])# first 3 crops
print(crops[1:5])#middle crops
print(crops[-5:-1])#middle crops excluding end crops
print(crops[-2:])# last two crops
print(crops[::1])#no change
print(crops[::2])# alternate crops
print(crops[::-1])#reversing the list
```

```python
list1=[10,20,30,40,50,60]
     # -6,-5,-4,-3,-2,-1
le=len(list1)
print(le)
# if i want till last element :6
print(list1[2:6],"1")
print(list1[-5:-1],"2")# 60 last element is not read here
print(list1[2:le],"3")# from 2 till complete
print(list1[::-1],"4")# reverses the list
print(list1[4:2:-1],"5")
print(list1[2:6])
print(list1[2:6])

```
