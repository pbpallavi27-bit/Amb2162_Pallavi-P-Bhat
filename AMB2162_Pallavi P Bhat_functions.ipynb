```python
# local variable in function
#global not only within function but also outside function it will be excecuted
# global is a key word to access a program
# once you enter global then that variable assigned value is permanent

def func1():
  global y
  print("Y is :",y)
  y=2
  print("changed y:",y)

y=50
func1()
print("value of y is:",y)
```
