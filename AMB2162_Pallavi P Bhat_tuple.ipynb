```python
# getting id number
# before and after will have the same memory space
# "1 number will occupy the 1 memory space "!!!!!"
l1=[1,2,3,4]
print(id(l1))
l1.append(21)
print(id(l1))
l1.insert(2,21)
print(id(l1))
l1.append(31)
print(id(l1))
```
