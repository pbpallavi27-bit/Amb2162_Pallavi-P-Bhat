```python
 import pandas as pd
df=pd.read_csv('Salary_Data.csv')
df
```

```python
# importing the necesary pacakages
# reading the dataset
# sclicing the data set as independent and dependent variables
# pre processing which is optional
# handling the missing data and identifiying outlyers
# splitting the data set into training and testing
#standardization and normalization which is preprocessing step and optional
# creating an object of an algorithm or the model
# train the model using fit method
# predict the model on the testing data set
# calculate intercept and qoueficient for linear regression
# evalutation , for regression finding different types of errors
# for classification calculating accuracy precision recall
# visualization
```

```python
# slice the dataset as descriptive (independent) and
# target variable (dependent)
x=df.iloc[:,0].values
y=df.iloc[:,-1].values
print(x)
print(y)

```

```python
# splitting the data set into training set and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
print(df.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# calculate the mean values of input and output
import numpy as np
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
print(x_mean,y_mean)
```

```python
#calculate intecept and coefficient
n=len(x_train)
num,den=0,0
for i in range(n):
  num+=(x_train[i]-x_mean)*(y_train[i]-y_mean)
  den+=(x_train[i]-x_mean)**2

b1=num/den
b0=y_mean-(b1*x_mean)
print("b0:",b0)
```

```python
#predicting the results on test dataset
y_pred_code=b0+b1*x_test
print(y_pred_code)
print()
print(y_test)
```

```python
# mean square error
n1=len(y_test)
nse=0
for i in range(n1):
  nse+=(y_test[i]-y_pred_code[i])**2
nse=nse/n1
print(nse)
```

```python
#root mean square error
rmse=np.sqrt(nse)
print(rmse)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train.reshape(-1, 1), y_train)
```

```python
print("intercept",reg.intercept_)
print("coeeficient",reg.coef_)
```

```python
y_pred = reg.predict(x_test.reshape(-1, 1))
#evaluation metrics
from sklearn import metrics
print("mean absolute error : ")
print(metrics.mean_absolute_error(y_test,y_pred))
print(metrics.mean_squared_error(y_test,y_pred))
print("root mean error :",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```python
import matplotlib.pyplot as plt
#visualisation
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_train,reg.predict(x_train.reshape(-1, 1)),color='orange')
plt.title('Salary vs Experience -Training set')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```
